```python
# Speed check

# list
a1 = [i for i in range(10000000)]
a2 = [i for i in range(10000000, 20000000)]

c = []

import time;
start = time.time()

for i in range(len(a1)):
    c.append(a1[i] + a2[i])

print(time.time() - start)

# numpy (size fixed & not referential array -> direct store value in memory not address)
b1 = np.arange(10000000)
b2 = np.arange(10000000, 20000000)

start1 = time.time()
c1 = b1 + b2
print(time.time() - start1)
```

```python
# Boolean indexing

# Random generate int
a = np.random.randint(1, 100, 24).reshape(6, 4)
a[a > 50]
a[a % 2 == 0]
a[(a > 50) & (a % 2 == 0)]
a[a % 7 != 0]
```

```python
# Broadcasting -> how numpy treats arrays with different shapes during arithmetic operations

''' 
1. if dimension different add 1 in head of small dimension
2. increase (1) until it same to the large dimension
3. if there is a dimension whose size is not 1, it can not be broadcasted
'''

# Same shape
a = np.arange(6).reshape(2, 3)
b = np.arange(6, 12).reshape(2, 3)

# print(a + b)

# Different shape
a = np.arange(6).reshape(2, 3)
b = np.arange(3)

# Example - 1
a = np.arange(12).reshape(4, 3)
b = np.arange(3).reshape(3)

print(a + b) # possible

# Example - 2
a = np.arange(12).reshape(3, 4)
b = np.arange(3).reshape(3)

# print(a+b) impossible

# Example - 3
a = np.arange(3).reshape(1, 3)
b = np.arange(3).reshape(3, 1)

print(a + b) # possible 

# Example - 4
a = np.arange(3).reshape(1, 3)
b = np.arange(4).reshape(4, 1)

print(a + b) # possible 

# Example - 5
a = np.arange(12).reshape(3, 4)
b = np.arange(12).reshape(4, 3)

# print(a + b) impossible
```

```python
# Own sigmoid

def sigmoid(arr):
    return 1 / (1 + np.exp(-arr))

a = np.arange(10)
sigmoid(a)
```

```python
# Own Mean Squared Error

def mean_squared_error(a, p):
    c = (a - p)**2
    return np.mean(c)

actual = np.random.randint(1, 50, 25)
predicted = np.random.randint(1, 50, 25)

mean_squared_error(actual, predicted)
```
